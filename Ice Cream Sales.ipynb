```python
df = pd.read_csv('./Monthly Ice Cream Shop Sales by Flavor.csv')

df
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x='Month', y='Total Units', marker='o', color='blue')

plt.xlabel('Month')
plt.ylabel('Total Units')
plt.title('Total Units Across Months')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Month', y='Total Profit', palette='viridis')

plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')

```

```python
df_long = pd.melt(df, id_vars='Month', var_name='Flavor', value_name='Sales')
df_long = df_long[df_long['Flavor'] != 'Total Profit']


plt.figure(figsize=(10, 6))
sns.lineplot(data=df_long, x='Month', y='Sales', hue='Flavor', marker='o')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales of Ice Cream Flavors Over Time')
```
